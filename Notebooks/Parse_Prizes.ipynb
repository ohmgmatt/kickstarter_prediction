```python
urlany = df.loc[212]['web_url']
r = urllib.request.urlopen(urlany)
if r.getcode() == 200:
    pass
else:
    print(r.getcode())
    exit()

container = BeautifulSoup(r.read()).find_all('div', 'pledge__info')
pledges = []
for item in container:
    for money in item.find_all('span', 'money'):
        cash = money.text
        i = re.search("[0-9]", cash)
        pledge = cash[i.start():]
        pledge = int(pledge.replace(',', ''))
        pledges.append(pledge)

print(pledges)
```
